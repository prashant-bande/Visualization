### Sample Plot using matplotlib, pandas

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('schoolimprovement2010grants.csv')
```

```python
df.head()
```

```python
# Display pandas histogram

df['Award_Amount'].plot.hist()
```

```python
# Display a Seaborn distplot

sns.distplot(df['Award_Amount'])
```

### Using the distribution plot

The distplot() function will return a Kernel Density Estimate (KDE) by default

```python
# with default options

sns.distplot(df['Award_Amount'])
```

```python
# custom bins

sns.distplot(df['Award_Amount'], kde=False, bins=20)
```

```python
# More options, vertical plot

sns.distplot(df['Award_Amount'], kde=False, bins=10, vertical=True, color='r')
```

```python
# with rug, color

sns.distplot(df['Award_Amount'], hist=False, rug=True, color='r')
```

```python
# with shade, rug

sns.distplot(df['Award_Amount'], hist=False, rug=True, kde_kws={'shade':True})
```

### Colors in Seaborn

Seaborn supports assigning colors to plots using matplotlib color codes

```python
sns.set(color_codes=True)
sns.distplot(df['premiums'], color='g')
```

Seaborn uses the set_palette() function to define a palette

```python
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    sns.distplot(df['premiums'])
```

##### Displaying Palettes


sns.palplot() function displays a palette

sns.color_palette() returns the current palette

```python
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    sns.palplot(sns.color_palette())
    plt.show()
```

##### Defining Custom Palettes

```python
# Circular colors = when the data is not ordered
sns.palplot(sns.color_palette("Paired", 12))
```

```python
# Sequential colors = when the data has a consistent range from high to low
sns.palplot(sns.color_palette("Blues", 12))
```

```python
# Diverging colors = when both the low and high values are interesting
sns.palplot(sns.color_palette("BrBG", 12))
```

```python
# Set style, enable color code, and create a magenta distplot

sns.set(color_codes=True)
sns.distplot(fmr['fmr_3'], color='magenta')
```

```python
# Loop through differences between bright and colorblind palettes
for p in ['bright', 'colorblind']:
    sns.set_palette(p)
    sns.distplot(fmr['fmr_3'])
    plt.show()
```

### Customizing with matplotlib


##### matplotlib Axes

Most customization available through matplotlib Axes objects

Axes can be passed to seaborn functions

```python
fig, ax = plt.subplots()
sns.distplot(df['premiums'], ax=ax)
ax.set(xlabel="Premium 2013-14")
```

#### Further Customizations

The axes object supports many common customizations


```python
fig, ax = plt.subplots()
sns.distplot(df['premiums'], ax=ax)
ax.set(xlabel="Premium 2013-14",
       ylabel="Distribution", xlim=(0, 1800), 
title="2013-14 Premium and Fees Distribution")
```

```python
##### Combining Plots

fig, (ax0, ax1) = plt.subplots(nrows=1,ncols=2,
                               sharey=True, figsize=(11,6))
sns.distplot(df['premiums'], ax=ax0)
sns.distplot(df.query('Region == "West"')['premiums'], ax=ax1)
ax1.set(xlabel="Premium (West)", xlim=(0, 1500))
ax1.axvline(x=850, label='My Budget', linestyle='--')
ax1.legend()
```

```python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot the distribution of data
sns.distplot(fmr['fmr_3'], ax=ax)

# Create a more descriptive x axis label
ax.set(xlabel="3 Bedroom Fair Market Rent")
```

### Regression Plots

### Matrix plots


##### heatmap

A heatmap is a common matrix plot that can be used to graphically summarize the relationship between two variables.

Seaborn's heatmap() function requires data to be in a grid format

pandas crosstab() is frequently used to manipulate the data

```python
df = pd.read_csv('daily_show_guests_cleaned.csv')
print(df.head(3))
pd_crosstab = pd.crosstab(df["Group"], df["YEAR"])
pd_crosstab
```

```python
# Plot a heatmap of the table
sns.heatmap(pd_crosstab)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```

```python
bike = pd.read_csv('bike_share.csv')

sns.heatmap(pd.crosstab(bike["mnth"], bike["weekday"], values=bike["total_rentals"], aggfunc='mean'))
```

##### Customize a heatmap

```python
# Plot a heatmap of the table with no color bar and using the BuGn palette
sns.heatmap(pd_crosstab, cbar=False, cmap="YlGn", linewidths=0.3)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```

##### Centering a heatmap


```python
sns.heatmap(pd_crosstab, cmap="YlGnBu", cbar=True, center=pd_crosstab.iloc[1, 4])
```

##### Plotting a correlation matrix


```python
sns.heatmap(bike.corr())
```

### Creating Plots on Data Aware Grids

##### JointGrid 

Seaborn's JointGrid combines univariate plots such as histograms, rug plots and kde plots with bivariate plots such as scatter and regression plots

```python
bike = pd.read_csv('bike_share.csv')
```

```python
# Build a JointGrid comparing humidity and total_rentals
sns.set_style("whitegrid")
g = sns.JointGrid(x="hum", y="total_rentals", data=bike, xlim=(0.1, 1.0)) 

g.plot(sns.regplot, sns.distplot)
```

##### jointplot

```python
# Create a jointplot similar to the JointGrid 
sns.jointplot(x="hum", y="total_rentals", kind='reg', data=bike)
```

```python
# Plot temp vs. total_rentals as a regression plot
sns.jointplot(x="temp", y="total_rentals", kind='reg', data=bike, order=2, xlim=(0, 1))
```

```python
# Plot a jointplot showing the residuals
sns.jointplot(x="temp", y="total_rentals", kind='resid', data=bike, order=2)
```

```python
# Create a jointplot of temp vs. casual riders
# Include a kdeplot over the scatter plot
g = (sns.jointplot(x="temp", y="casual", kind='scatter', data=bike, marginal_kws=dict(bins=10, rug=True)) .plot_joint(sns.kdeplot))
```

```python
# Replicate the above plot but only for registered riders
g = (sns.jointplot(x="temp",
             y="registered",
             kind='scatter',
             data=bike,
             marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))
```
