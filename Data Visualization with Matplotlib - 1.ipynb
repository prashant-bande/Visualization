## Basic Plots

### Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Set up figure size and DPI for screen
plt.rcParams['figure.figsize'] = (12,9)
plt.rcParams['figure.dpi'] = 50
plt.rcParams['font.size'] = 16
```

```python
xval = np.arange(0,10,0.1)
```

```python
plt.plot(xval, np.sin(xval))

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
# Multiple plots
plt.plot(xval, np.sin(xval))
plt.plot(xval, np.cos(xval))

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
# legend and Title

plt.plot(xval, np.sin(xval), label="y=sin(x)")
plt.plot(xval, np.cos(xval), label=r"y=cos(x)")
plt.legend()
plt.xlabel('X')
plt.ylabel(r'Y')
```

```python
# Line Color

plt.plot(xval, np.sin(xval), label="y=sin(x)", color='g')
plt.plot(xval, np.cos(xval), label=r"y=cos(x)", color='r')
plt.legend()
plt.xlabel('X')
plt.ylabel(r'Y')
```

```python
# Line Style

plt.plot(xval, np.sin(xval), label="y=sin(x)", color='g', linestyle='--')
plt.plot(xval, np.cos(xval), label=r"y=cos(x)", color='r', marker='o', linestyle='none')
plt.legend()
plt.xlabel('X')
plt.ylabel(r'Y')
```

```python
# short format code

plt.plot(xval, np.sin(xval), 'ro--', label="y=sin(x)")
plt.plot(xval, np.cos(xval), 'b.-.', label=r"y=cos(x)")
plt.legend()
plt.xlabel('X')
plt.ylabel(r'Y')
```

```python
# Axis limit

plt.plot(xval, np.sin(xval), 'g+--', label='y=sin(x)', ms=30)
plt.plot(xval, np.cos(xval), label='y=cos(x)')
plt.ylim(1,0)
plt.xlim(0,1)
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

### Histogram

```python
# Gen 1000 random numbers
rands = np.random.normal(size=1000)
plt.hist(rands)
plt.xlabel('X')
plt.ylabel('Pr(x)')
```

```python
# Horizontal

rands = np.random.normal(size=1000)
plt.hist(rands, orientation='horizontal')
plt.ylabel('X')
plt.xlabel('Pr(x)')
plt.title('Horizontal Histogram')
```

```python
# Different line style
idx = 0
for marker in ('-', '--', '-.', ':'):
    plt.plot(xval, np.sin(xval)+idx, ''.join(('k', marker)), label=marker)
    idx += 1

plt.legend(loc=9)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Style')
```

```python
# Marker frequency with markevery

xvals2 = np.arange(0, 10, 0.01)
plt.plot(xvals2, np.sin(xvals2), 'k.', markevery=10, ms=15)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# Different markers

idx = 0
for marker in ('.', ',', 'o', 'v', '^', '>', '<', '1', '2', '3', '4', 's', 'p', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_'):
    plt.plot(idx, marker, ms=15)
    idx += 1
```

```python
idx = 0
for marker in ('.', ',', 'o', 'v', '^', '>', '<', '1', '2', '3', '4', 's', 'p', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_'):
    plt.plot(idx, 0, ''.join(('r', marker)), ms=15)
    idx += 1
```

```python
# Line width

plt.plot(xval, np.sin(xval), linewidth=10)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# zorder - Larger value closer to the graph

plt.plot(xval, np.sin(xval), linewidth=10, zorder=1)
plt.plot(xval+1, np.sin(xval), linewidth=10, zorder=0)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# Transparency

plt.plot(xval, np.sin(xval), linewidth=10, alpha=0.3)
plt.plot(xval+1, np.sin(xval), linewidth=10, alpha=0.3)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# Array order matters!

shuffled_xval = np.random.permutation(xval)
plt.plot(shuffled_xval, np.sin(shuffled_xval), 'k.')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

plt.plot(shuffled_xval, np.sin(shuffled_xval), 'k-')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### scatter()

```python
# Basic scatter plot

xval = np.arange(0, 10, 0.1)
plt.scatter(xval, np.sin(xval))
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# color

xval = np.arange(0, 10, 0.1)
plt.scatter(xval, np.sin(xval), c=np.cos(xval))
plt.colorbar()
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# color map and marker

xval = np.arange(0, 10, 0.1)
plt.scatter(xval, np.sin(xval), c=np.cos(xval), cmap='inferno', marker='*')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# size and edgecolors

xval = np.arange(0, 10, 0.1)
plt.scatter(xval, np.sin(xval), c=np.cos(xval), cmap='viridis', s=np.power(xval, 3), edgecolors='w')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('sin(x)')
```

## Uncertainties of Line and Scattered plot

```python
nums = np.arange(0,10,0.5)
xerrors = np.random.uniform(size=20)
yerrors = np.random.uniform(size=20)
```

```python
plt.plot(nums, np.sin(nums))
plt.show()
plt.errorbar(nums, np.sin(nums), xerr=xerrors, yerr=yerrors)
```

```python
# Errorbar color and elinewidth

plt.errorbar(nums, np.sin(nums), xerr=xerrors, yerr=yerrors, elinewidth=2, ecolor='red')
```

```python
plt.errorbar(nums, np.sin(nums), xerr=xerrors, yerr=yerrors, capsize=4)
```

```python
# Uncertainties on bar plot

plt.bar(nums, np.sin(nums)**2, width=0.3, xerr=xerrors, yerr=yerrors)
```

### Transformation

```python
# transData (data -> display)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(1,1, 'ro')
print(ax.transData.transform((1,1)))
plt.clf()
```

```python
# transAxes (axes -> display)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(1,1, 'ro')
print(ax.transAxes.transform((1,1)))
plt.clf()
```

```python
# transFigure (figure -> display)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(1,1, 'ro')
print(fig.transFigure.transform((1,1)))
plt.clf()
```

```python
# invert transformation (display -> data)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(1,1, 'ro')
inv = ax.transData.inverted()
print(inv.transform((300,225)))
plt.clf()
```

### using formatter



```python
# NullFormatter()
plt.plot()
plt.gca().yaxis.set_major_formatter(ticker.NullFormatter())
```

```python
# Log vs Linear formatting

plt.loglog()
plt.ylim(1, 1e7)
```

```python
# EngFormatter w/ unit

plt.loglog()
plt.ylim(1, 10000)
plt.gca().yaxis.set_major_formatter(ticker.EngFormatter(unit='Hz'))
```

```python
# FormatStrFormatter

plt.plot()
plt.gca().yaxis.set_major_formatter(ticker.FormatStrFormatter('Value %3.2f'))
```

```python
# FuncFormatter

func = lambda x,pos: r"$\sqrt{\rm tick}$ is %3.2f" %np.sqrt(x)
plt.plot()
plt.ylim(1,10)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(func))
```
